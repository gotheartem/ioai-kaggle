```python
import tqdm
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/raw/train.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'].apply(lambda x: x[:-2] + '01'), format='%Y-%m-%d')
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
gb = df.groupby(['Year', 'Month', 'Company_ID', 'Product_ID'])
mode = lambda x: np.nan if x.isna().all() else x.mode()
ndf = pd.concat([
    gb['Date'].first().reset_index()['Date'], 
    gb['Target'].first().reset_index()['Company_ID'], 
    gb['Target'].first().reset_index()['Product_ID'], 

    gb['cat_0'].agg(mode).reset_index()['cat_0'], 
    gb['cat_1'].agg(mode).reset_index()['cat_1'], 
    gb['cat_2'].agg(mode).reset_index()['cat_2'], 
    gb['cat_3'].agg(mode).reset_index()['cat_3'], 
    gb['cat_4'].agg(mode).reset_index()['cat_4'], 
    gb['cat_5'].agg(mode).reset_index()['cat_5'], 
    gb['cat_6'].agg(mode).reset_index()['cat_6'], 
    gb['cat_7'].agg(mode).reset_index()['cat_7'], 
    gb['cat_8'].agg(mode).reset_index()['cat_8'], 
    gb['cat_9'].agg(mode).reset_index()['cat_9'], 
    gb['cat_10'].agg(mode).reset_index()['cat_10'], 
    gb['cat_11'].agg(mode).reset_index()['cat_11'], 
    gb['cat_12'].agg(mode).reset_index()['cat_12'], 
    gb['cat_13'].agg(mode).reset_index()['cat_13'], 
    gb['cat_14'].agg(mode).reset_index()['cat_14'], 

    gb['num_0'].agg(mode).reset_index()['num_0'], 
    gb['num_1'].agg(mode).reset_index()['num_1'], 
    gb['num_2'].agg(mode).reset_index()['num_2'], 
    gb['num_3'].agg(mode).reset_index()['num_3'], 
    gb['num_4'].agg(mode).reset_index()['num_4'], 
    gb['num_5'].agg(mode).reset_index()['num_5'], 
    gb['num_6'].agg(mode).reset_index()['num_6'], 
    gb['num_7'].agg(mode).reset_index()['num_7'], 
    gb['num_8'].agg(mode).reset_index()['num_8'], 
    gb['num_9'].agg(mode).reset_index()['num_9'], 
    gb['num_11'].agg(mode).reset_index()['num_11'], 
    gb['num_12'].agg(mode).reset_index()['num_12'], 

    gb['num_10'].median().reset_index()['num_10'], 
    gb['num_13'].median().reset_index()['num_13'], 
    gb['num_14'].median().reset_index()['num_14'], 
    gb['num_15'].median().reset_index()['num_15'], 
    gb['num_16'].median().reset_index()['num_16'], 
    gb['num_17'].median().reset_index()['num_17'], 
    gb['num_18'].median().reset_index()['num_18'], 

    gb['Target'].sum().reset_index()['Target']
], axis=1)
```

```python
ndf
```

```python
features = ['cat_0', 'cat_1', 'cat_2', 'cat_3', 'cat_4', 'cat_5', 'cat_6', 'cat_7', 'cat_8', 'cat_9', 'cat_10', 'cat_11', 'cat_12', 'cat_13', 'cat_14', 
            'num_0', 'num_1', 'num_2', 'num_3', 'num_4', 'num_5', 'num_6', 'num_7', 'num_8', 'num_9', 'num_11', 'num_12']
for feat in tqdm.tqdm(features):
    cvc = df[feat].value_counts()
    ndf[feat] = ndf[feat].apply(lambda x: (x if np.array(x).ndim == 0 else cvc[x].sort_values(ascending=False).index[0]))
```
